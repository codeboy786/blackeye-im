#**1. Data Preparation** <font color = red>[5 marks]</font> <br>

Utilise the pandas method to read tables from the html document
and extract the requried data consists of a structured format.

```python
# Initialize Spark session

# URL of the Wikipedia page
url = 'https://web.archive.org/web/20230908091635/https://en.wikipedia.org/wiki/List_of_largest_banks'

# Use pandas to read all HTML tables from the page
tables = pd.read_html(url)

# Load the correct table (Table 1)
df_pandas = tables[1]

# Columns should be: Rank, Bank name, Market cap (US$ billion)
df_pandas = df_pandas[['Rank', 'Bank name', 'Market cap (US$ billion)']]

# Convert pandas DataFrame to PySpark DataFrame
bank_df = spark.createDataFrame(df_pandas)

# Rename columns to remove spaces and special characters for easier processing
bank_df = bank_df.withColumnRenamed("Rank", "Rank") \
                 .withColumnRenamed("Bank name", "Bank_Name") \
                 .withColumnRenamed("Market cap (US$ billion)", "Market_Cap_USD_Billion")

# Show the first few rows of the PySpark DataFrame
print("Top 5 rows of the dataframe:")
bank_df.show(5)

# Print the schema of the DataFrame to check data types
print("Dataframe Schema:")
bank_df.printSchema()
```

#**2. Data Cleaning** <font color = red>[20 marks]</font> <br>

##**2.1 Handle Missing Values** <font color = red>[10 marks]</font> <br>


```python
# Print the schema to check data types
print("Current Schema:")
bank_df.printSchema()
```

```python
# Check for missing values in each column
from pyspark.sql.functions import col, sum as spark_sum

print("Missing values per column:")
# We verify if any column has null values
for column in bank_df.columns:
    # We count the number of rows where the specific column is Null
    missing_count = bank_df.filter(col(column).isNull()).count()
    print(f"{column}: {missing_count}")
```

```python
# Drop rows with missing values
# If any row has missing data, we will remove it to maintain data integrity.
bank_df_clean = bank_df.na.drop()

# Verify the count after cleaning
original_count = bank_df.count()
clean_count = bank_df_clean.count()

print(f"Original Row Count: {original_count}")
print(f"Cleaned Row Count: {clean_count}")

# Replace the original dataframe with the cleaned one for future steps
bank_df = bank_df_clean
```

##**2.2 Fixing Columns** <font color = red>[5 marks]</font> <br>


The Market cap is already in numeric so no conversion needed.

```python
# First, let's ensure the Market Cap column is definitely numeric (Double type)
# Even if it looks like a number, sometimes Spark reads it as a string initially.
from pyspark.sql.types import DoubleType

bank_df = bank_df.withColumn("Market_Cap_USD_Billion", col("Market_Cap_USD_Billion").cast(DoubleType()))

# Count the total number of rows
total_rows = bank_df.count()
print(f"Total number of banks: {total_rows}")

# Check for duplicates
# We group by Bank_Name and count how many times each appears.
duplicate_count = bank_df.groupBy("Bank_Name").count().filter(col("count") > 1).count()

print(f"Number of duplicate rows found: {duplicate_count}")
```

##**2.2 Handle Outliers** <font color = red>[5 marks]</font> <br>


```python
# Write code for outlier analysis
# Calculate basic statistics to identify outliers using the IQR method
quantiles = bank_df.approxQuantile("Market_Cap_USD_Billion", [0.25, 0.75], 0.05)
Q1 = quantiles[0]
Q3 = quantiles[1]
IQR = Q3 - Q1

# Define the bounds for what is considered an outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Filter the dataframe to see which banks are considered outliers
outliers_df = bank_df.filter((col("Market_Cap_USD_Billion") < lower_bound) |
                             (col("Market_Cap_USD_Billion") > upper_bound))

print("Outliers identified (Banks with exceptionally high Market Cap):")
outliers_df.show()
```

#**4. ETL and Querying** <font color = red>[45 marks]</font> <br>

```python
#Imports & Spark Session Initialization

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, round
import logging
from datetime import datetime
import sqlite3
import pandas as pd

# Initialize Spark Session
spark = SparkSession.builder \
    .appName("BankingDataAnalysis") \
    .getOrCreate()

# Logging Configuration
logfile = "code_log.txt"
```

```python
#Logging Function

def log_progress(msg):
    timeformat = '%Y-%h-%d-%H:%M:%S'
    timestamp = datetime.now().strftime(timeformat)
    with open(logfile, 'a') as f:
        f.write(f"{timestamp} : {msg}\n")
```

```python
# Import required functions
from pyspark.sql.functions import col, avg, sum, count, desc, asc, round, lag, dense_rank, ntile, when, first, lead
from pyspark.sql.window import Window

# ETL STEP: TRANSFORM (Currency Conversion)
# Collect exchange rates
# We assume exchange_rate_df was loaded earlier.
rates = exchange_rate_df.collect()
rate_dict = {row['Currency']: row['Rate'] for row in rates}

gbp_rate = rate_dict.get('GBP', 1.0)
eur_rate = rate_dict.get('EUR', 1.0)
inr_rate = rate_dict.get('INR', 1.0)

# Apply conversion to create new columns
bank_df_transformed = bank_df.withColumn("Market_Cap_GBP", round(col("Market_Cap_USD_Billion") * gbp_rate, 2)) \
                             .withColumn("Market_Cap_EUR", round(col("Market_Cap_USD_Billion") * eur_rate, 2)) \
                             .withColumn("Market_Cap_INR", round(col("Market_Cap_USD_Billion") * inr_rate, 2))

# ETL STEP: LOAD (Create View for SQL)
# Register the DataFrame as a temporary view so we can run SQL queries on it
bank_df_transformed.createOrReplaceTempView("largest_banks")

# Verify it worked
print("Data Transformed and Temp View 'largest_banks' created.")
bank_df_transformed.show(3)
```

##**4.1. Market Capitalization Analysis** <font color = red>[3 marks]</font> <br>

Q: Perform Advanced Market Capitalization Analysis with Growth Metrics

```python
# Query: Advanced Market Cap Analysis with Growth Metrics
# We analyze the share of each bank relative to the total market cap.

# Calculate the Total Market Cap of all banks
total_market_cap = bank_df_transformed.select(sum("Market_Cap_USD_Billion")).collect()[0][0]

# Add a column for Market Share Percentage
market_analysis_df = bank_df_transformed.withColumn("Market_Share_Pct",
                                                    round((col("Market_Cap_USD_Billion") / total_market_cap) * 100, 2))

# Display Top 5 Banks with their calculated Market Share
print("Top 5 Banks - Market Share Analysis:")
market_analysis_df.select("Rank", "Bank_Name", "Market_Cap_USD_Billion", "Market_Share_Pct").show(5)
```

##4.2.**Market Concentration Analysis** <font color = red>[6 marks]</font> <br>

Q: Analyze Market Concentration and Categorize Banks Based on Market Share Tiers

```python
# Market Concentration Analysis
# We categorize banks into tiers based on their Rank.
# Tier 1: Top 10 Banks (Dominant)
# Tier 2: Rank 11-50 (Mid-tier)
# Tier 3: Rank 51+ (Small-tier)

tier_df = bank_df_transformed.withColumn("Tier",
                                         when(col("Rank") <= 10, "Tier 1 (Top 10)")
                                         .when((col("Rank") > 10) & (col("Rank") <= 50), "Tier 2 (Mid-Size)")
                                         .otherwise("Tier 3 (Small-Size)"))

# Group by Tier to see the concentration of wealth in each tier
tier_stats = tier_df.groupBy("Tier").agg(
    count("Bank_Name").alias("Num_Banks"),
    sum("Market_Cap_USD_Billion").alias("Total_Market_Cap"),
    avg("Market_Cap_USD_Billion").alias("Avg_Market_Cap")
).orderBy("Total_Market_Cap", ascending=False)

print("Market Concentration by Tiers:")
tier_stats.show(truncate=False)
```

##**4.3. Market Capitalization Distribution** <font color = red>[6 marks]</font> <br>

Q: Examine Statistical Distribution of Market Capitalization Using Quartile Analysis

```python
# Import required functions
from pyspark.sql.functions import (col, avg, sum, count, desc, asc, round, lag,
                                 dense_rank, ntile, when, first, lead, min, max, lit)
from pyspark.sql.window import Window

# Query 3: Statistical Distribution Analysis
# We use the ntile() window function to divide banks into 4 groups (Quartiles)
# Quartile 1 = Smallest banks in the list
# Quartile 4 = Largest banks in the list

window_spec = Window.orderBy("Market_Cap_USD_Billion")

quartile_df = bank_df_transformed.withColumn("Quartile", ntile(4).over(window_spec))

# Analyze the average market cap per quartile
quartile_analysis = quartile_df.groupBy("Quartile").agg(
    min("Market_Cap_USD_Billion").alias("Min_Cap"),
    max("Market_Cap_USD_Billion").alias("Max_Cap"),
    avg("Market_Cap_USD_Billion").alias("Avg_Cap"),
    count("Bank_Name").alias("Count")
).orderBy("Quartile")

print("Market Cap Distribution by Quartiles:")
quartile_analysis.show()
```

##**4.4. Comparative Size Analysis** <font color = red>[6 marks]</font> <br>

Q: Conduct Comparative Size Analysis to Classify Banks by Relative Market Size

```python
# Comparative Size Analysis
# Compare each bank's size against the global average market cap.
# Logic:
# - Large: > 2x the average
# - Medium: Between 0.5x and 2x the average
# - Small: < 0.5x the average

# First, get the average market cap
avg_cap = bank_df_transformed.select(avg("Market_Cap_USD_Billion")).collect()[0][0]
print(f"Global Average Market Cap: ${avg_cap:.2f} Billion")

# Now classify each bank
size_df = bank_df_transformed.withColumn("Relative_Size",
                                         when(col("Market_Cap_USD_Billion") >= (2 * avg_cap), "Large")
                                         .when(col("Market_Cap_USD_Billion") < (0.5 * avg_cap), "Small")
                                         .otherwise("Medium"))

# Show distribution of these classes
print("Banks classified by Relative Size:")
size_df.groupBy("Relative_Size").count().show()
```

##**4.5. Market Growth Analysis** <font color = red>[6 marks]</font> <br>

Q: Evaluate Market Growth and Identify Gaps Between Consecutive Banks

```python
# Growth and Gap Analysis
# Define window ordered by Rank
rank_window = Window.orderBy("Rank")

# Calculate the gap: (Current Bank Cap) - (Next Bank Cap)
# We use lead() to look at the 'next' bank in the rank list (Rank 1 vs Rank 2)
gap_df = bank_df_transformed.withColumn("Next_Bank_Cap", lead("Market_Cap_USD_Billion").over(rank_window)) \
                            .withColumn("Gap_USD_Billion", col("Market_Cap_USD_Billion") - col("Next_Bank_Cap"))

print("Top 5 Gaps between consecutive banks (Where competition is steepest):")
gap_df.select("Rank", "Bank_Name", "Market_Cap_USD_Billion", "Next_Bank_Cap", "Gap_USD_Billion").show(5)
```

##**4.6. Market Dominance Analysis** <font color = red>[6 marks]</font> <br>

Q: Assess Market Dominance by Measuring Cumulative Share and Dominance Score

```python
# Market Dominance Analysis
# We calculate the running total of market share to see how concentrated the industry is.

window_desc = Window.orderBy(desc("Market_Cap_USD_Billion"))

# Calculate cumulative sum of market cap
dominance_df = bank_df_transformed.withColumn("Cumulative_Cap", sum("Market_Cap_USD_Billion").over(window_desc))

# Calculate total market cap (scalar value)
total_cap = bank_df_transformed.select(sum("Market_Cap_USD_Billion")).collect()[0][0]

# Calculate Cumulative Share %
dominance_df = dominance_df.withColumn("Cumulative_Share_Pct",
                                       round((col("Cumulative_Cap") / total_cap) * 100, 2))

print("Market Dominance (Cumulative Share of Top Banks):")
dominance_df.select("Rank", "Bank_Name", "Market_Cap_USD_Billion", "Cumulative_Share_Pct").show(10)
```

##**4.7. Segment-Wise Bank Analysis** <font color = red>[6 marks]</font> <br>

Q: Analyze Segment-Wise Bank Performance Based on Market Capitalization Ranges

```python
# Segment Performance Analysis
# We define specific ranges for Market Cap to analyze performance segments.

segment_df = bank_df_transformed.withColumn("Segment",
                                            when(col("Market_Cap_USD_Billion") < 100, "Under $100B")
                                            .when((col("Market_Cap_USD_Billion") >= 100) & (col("Market_Cap_USD_Billion") < 200), "$100B - $200B")
                                            .when((col("Market_Cap_USD_Billion") >= 200) & (col("Market_Cap_USD_Billion") < 300), "$200B - $300B")
                                            .otherwise("Over $300B"))

# Aggregate data by Segment
segment_stats = segment_df.groupBy("Segment").agg(
    count("Bank_Name").alias("Bank_Count"),
    avg("Market_Cap_USD_Billion").alias("Avg_Market_Cap")
).orderBy("Avg_Market_Cap", ascending=False)

print("Performance by Market Cap Segment:")
segment_stats.show()
```

##**4.8. Performance Dashboard** <font color = red>[6 marks]</font> <br>

Q: Generate a Comprehensive Performance Dashboard for Bank Rankings and Metrics

```python
# Comprehensive Performance Dashboard
# We join multiple metrics into a single view for the final report.

# We start with the base transformed dataframe (includes Currencies)
# We add the "Tier" column we defined in 4.2
dashboard_df = bank_df_transformed.withColumn("Tier",
                                         when(col("Rank") <= 10, "Tier 1 (Top 10)")
                                         .when((col("Rank") > 10) & (col("Rank") <= 50), "Tier 2 (Mid-Size)")
                                         .otherwise("Tier 3 (Small-Size)"))

# We add the "Market Share" column from 4.1
total_cap = bank_df_transformed.select(sum("Market_Cap_USD_Billion")).collect()[0][0]
dashboard_df = dashboard_df.withColumn("Global_Market_Share_Pct",
                                       round((col("Market_Cap_USD_Billion") / total_cap) * 100, 2))

# Select the final columns for the dashboard
final_dashboard = dashboard_df.select(
    "Rank",
    "Bank_Name",
    "Tier",
    "Market_Cap_USD_Billion",
    "Market_Cap_GBP",
    "Market_Cap_EUR",
    "Market_Cap_INR",
    "Global_Market_Share_Pct"
).orderBy("Rank")

print("Final Comprehensive Dashboard (First 10 Rows):")
final_dashboard.show(10)

# Optional: Save this final dashboard to CSV for the "Visualization Integration" task
final_dashboard.write.csv("Final_Bank_Dashboard", header=True, mode="overwrite")
print("Dashboard data saved to 'Final_Bank_Dashboard' folder.")
```
